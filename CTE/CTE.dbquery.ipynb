```sql
WITH CTE_Total_Sales AS (
  SELECT
    CustomerID,
    SUM(SaleAmount) AS Total
  FROM
    orders
  GROUP BY
    CustomerID
  ORDER BY
    Total DESC
),

CTE_LAST_ORDER AS (
    SELECT
      CustomerID, 
      MAX(OrderDate)LOD from orders
  GROUP BY CustomerID
)

select 
c.CustomerID,
c.CustomerName,
c.City,
t.total,
L.LOD
from customers as c 
left join CTE_Total_Sales as t 
on c.CustomerID = t.CustomerID
left join CTE_LAST_ORDER as L
on c.CustomerID = L.CustomerID;
```
