```sql
select * from workspace.sarang.bookstore_data

#### using From clause 
SELECT
  Book_ID,
  Price
FROM (
  SELECT 
    Book_ID,
    Price, 
    AVG(Price) OVER () AS avg_price 
  FROM workspace.sarang.bookstore_data
) t
WHERE Price > t.avg_price

##### Using Where clause 
SELECT 
  Book_ID,
  Price
FROM workspace.sarang.bookstore_data
WHERE Price > (
  SELECT AVG(Price)
  FROM workspace.sarang.bookstore_data
)

_______________________________________________________

select *, rank() over (order by total_sales desc) as rnk from (
SELECT Customer_ID, sum(Sale_Amount) total_sales from workspace.sarang.customer_sales group by Customer_ID order by total_sales desc)

```
