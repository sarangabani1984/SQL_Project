```sql
-- Aggregate sales data by country
select
  country,                                         -- Country name
  count(1) as total_sales,                         -- Total number of sales records
  avg(sale_amount) as avg_sale_amount,             -- Average sale amount
  sum(sale_amount) as total_sale_amount,           -- Total sale amount
  max(sale_amount) as max_sale_amount,             -- Maximum sale amount
  min(sale_amount) as min_sale_amount              -- Minimum sale amount
from workspace.sarang.sales_data1
group by country

-- Calculate total sale amount per country using window function
select
  country,                                         -- Country name
  sum(sale_amount) over(partition by country) as total_sale_amount  -- Total sale amount per country
from workspace.sarang.sales_data1
```
